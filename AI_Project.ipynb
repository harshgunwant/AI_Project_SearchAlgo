```python
import matplotlib.pyplot as plt

data1 = [[0,0,0],[2,10,5],[4,24,33],[8,325,492],[12,2187,3642],[16,533,859],[20, 39768, 96415],[24, 62354, 301375]]
data2 = [[0,0,0],[2,4,2],[4,16,12],[8,325, 492],[12, 533, 859],[16, 2851, 4739],[20, 9112,16319],[24, 43427, 109154]]
data3 = [[0,0,0],[2,4,2],[4,15,9],[8,41,53],[12, 185, 254],[16, 730, 1146],[20, 2270, 3826],[24, 11440, 21105]]

x = [row[0] for row in data1]  # Extract x values from data1 (all datasets have the same x values)
y1 = [row[1] for row in data1]  # Extract y values from data1
y2 = [row[1] for row in data2]  # Extract y values from data2
y3 = [row[1] for row in data3]  # Extract y values from data3

plt.plot(x, y1, label='Uniform Cost Search')
plt.plot(x, y2, label='A* with Misplaced Tile Heuristic')
plt.plot(x, y3, label='A* with Manhattan Distance')

plt.title('Nodes Expanded vs Solution Depth')
plt.xlabel('Solution Depth')
plt.ylabel('Nodes Expanded')
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

data1 = [[11, 0], [15, 2], [18, 4], [8, 8], [82, 12], [82, 16], [266, 20], [356, 24]]
data2 = [[7, 0], [14, 2], [18, 4], [9, 8], [58, 12], [42, 16], [180, 20], [217, 24]]
data3 = [[9, 0], [14, 2], [15, 4], [8, 8], [39, 12], [30, 16], [113, 20], [95, 24]]

# Extracting the time and depth values from the data
time1, depth1 = zip(*data1)
time2, depth2 = zip(*data2)
time3, depth3 = zip(*data3)

# Plotting the graph
plt.plot(depth1, time1, label="Uniform Cost Search")
plt.plot(depth2, time2, label="A* with Misplaced Tile Heuristic")
plt.plot(depth3, time3, label="A* with Manhattan Distance")

# Adding labels and title to the graph
plt.xlabel("Depth")
plt.ylabel("Time Taken(in milliseconds)")
plt.title("Time Taken vs. Depth")
plt.legend()
plt.grid()
# Displaying the graph
plt.show()
```
